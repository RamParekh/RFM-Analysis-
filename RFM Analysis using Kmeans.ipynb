```python
# df['Price']=df['Price'].str.replace(",",".")
df['Price']= df['Price'].astype('float64')
```

```python
#Recency

day= '2012-01-01'
day = pd.to_datetime(day)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

recency = df.groupby(['CustomerID']).aggregate({'InvoiceDate': lambda x:(day-x.max()).days})
```

```python
recency.head()
```

```python
# frequency 

freq = df.drop_duplicates(subset='fBillno').groupby(['CustomerID'])[['fBillno']].count()
```

```python
freq.head()
```

```python
#monetary values

df['total'] = df['Price']*df['Quantity']
```

```python
money = df.groupby(['CustomerID'])[['total']].sum()
```

```python
money
```

```python
RFM = pd.concat([recency,freq,money], axis=1)
recency.columns=['Recency']
freq.columns = ['Frequency']
money.columns=['Monetary']
```

```python
RFM
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(RFM)
```

```python
RFM
```

```python
from sklearn.cluster import KMeans

inertia=[]

for i in np.arange(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(scaled)
    inertia.append(kmeans.inertia_)


plt.figure(figsize=(12,8))
plt.plot(inertia, marker="o")
```

```python
RFM
```

```python
kmeans= KMeans(n_clusters=3)
kmeans.fit(scaled)
RFM['Clusters'] = (kmeans.labels_ + 1)
```

```python
RFM
```

```python
group_RFM = RFM.groupby(['Clusters'])['Recency','Frequency','Monetary'].mean()
```

```python
group_RFM
```

```python
def func(row):
    if row['Clusters']==1:
        return 'avg customers'
    elif row['Clusters']==2:
        return 'lapsed customers'
    else:
        return 'Best Customers'
```

```python
RFM['Conditions'] = RFM.apply(func,axis=1)
```

```python
RFM
```

```python
result = RFM['Conditions'].value_counts()
```
